```python
import pandas as pd
import os
df = []
folderpath = "Listings/" #folder Listings contains one file for every cith with their listings (site: https://insideairbnb.com/get-the-data/)
for i in os.listdir(folderpath): #reading each file seperatly 
    city_df = pd.read_csv(os.path.join(folderpath,i))
    city_name = os.path.basename(i).replace(".csv", "").capitalize()
    city_df['City'] = city_name 
    df.append(city_df)
#merge the data for every city in one data frame
dfr = pd.concat(df, ignore_index = True) 
dfr
```

```python
## 1st: Finf the number of AirBnB listings per city
df_listings_per_city = dfr.groupby('City').size().reset_index(name='Listings')
df_listings_per_city
```

The df_listings_per_city data frame indicates the number of Airbnb listings across selected European cities. From the data frame we can see that the number of listings varies significantly across these cities, some of them like London and Paris having a very high number of Airbnb listings, while others, such as Dublin and Venice, have fewer. This of course depends mostly on how popular every city is with tourists and what the demand for accommodation needs are in each one of them. As we can observe London has the largest number of Airbnb listings with 96,182 listings. Εven though London has almost three times the number of listings that Madrid with 26,980 and Rome with 34,061 listings have, these two are in a medium range among tsose displayed cities. A significant difference can be observed in Dublin which has the fewest listings among the cities in the dataset with only 6,175 listings, a fact that could reflect either at a lower demand for short term rentals or a non peferable accomondation.

```python
## 2nd:Fing the AirBnB Densities per city

#File with populations foe every city from www.kaggle.com 
#(https://www.kaggle.com/datasets/rathiankit/european-city-population-and-area?resource=download)
population_df = pd.read_excel('Cities_Populations.xlsx') 

# Filtering population_df to include only the specific cities 
filtered_population_df = population_df[population_df['City'].isin(df_listings_per_city['City'])]

Densities_df = pd.merge(df_listings_per_city, filtered_population_df, on='City', how='inner')
# density: number of listings per 1,000 inhabitants
Densities_df['Density'] = (Densities_df['Listings'] / Densities_df['Population']) * 1000
Cities_df = Densities_df[['City', 'Listings', 'Population', 'Density']]
Cities_df
```

This analysis examines the density of Airbnb listings across 12 cities. By calculating Airbnb density we can see how prominent Airbnb rentals are within each city relative to its population size and also we have the opportunity to observe the availability of the market. The data frame Cities_df indicates the listings, the population and the destiny which shows how many Airbnb listings are available per 1,000 inhabitants in each city.
I found the population for each city from kaggle (https://www.kaggle.com/datasets/rathiankit/european-city-population-and-area?resource=download) a site which provides many data sets for analytics. The file 'Cities_Populations.xlsx' shows the population for each of the cities and its a very accurate source od data. 
In the analysis we can see that cities with high Airbnb density are Lisbon which has the highest , with 47.96 listings per 1.000 residents, followed by Paris at 43.26 . This high density suggests a significant reliance on Airbnb properties, likely due to high tourist activity and demand for short term stays. On the other side Berlin has a low Airbnb density with 3.71 , which could indicate stricter a large resident population, or lower demand for Airbnb accommodations relative to the population. An important report that I should mention is about London with 10.90 which has lower densities than might be expected for a major capital, possibly due to its large population.

```python
##3rd: AirBnB bookings and income per listing for each city

dfr['Estimated_Bookings'] = (dfr['number_of_reviews_ltm'] * 2) 
dfr['Estimated_Nights'] = dfr['Estimated_Bookings'] * 3 
dfr['Income_per_Listing'] = dfr['Estimated_Nights'] * dfr['price'] #income per listing
dfr[['id', 'name', 'Estimated_Nights', 'Income_per_Listing']]
```

```python
#average income per listing for each city
average_income_per_city = dfr.groupby('City')['Income_per_Listing'].mean().reset_index()
#average nights per listing for each city
average_nights_per_city = dfr.groupby('City')['Estimated_Nights'].mean().reset_index()
city_nights_income = pd.merge(average_income_per_city, average_nights_per_city, on='City', how='inner')
city_nights_income.rename(columns={'Income_per_Listing': 'Average_Income','Estimated_Nights': 'Average_Nights' }, inplace=True)
Cities_df = Cities_df.merge(city_nights_income, on='City', how='left')
city_nights_income
```

This data frame city_nights_income, displayes the average income and estimated nights booked per listing across these European cities. With those metrics we can understand the potential revenue and booking activity in different locations, something which can really help Airbnb hosts and investors by providing valuable insights for growing their revenue.
Copenhagen stands out with a very high average income per listing at 50,349.8€ likely influenced by high nightly rates or limited availability, leading to increased competition. Venice also report really high average income up to 26,046.42, something that can be due to strong demand, as it is a hight tourist accomodation, allowing hosts to charge premium prices.
Cities with lowest average income are London and Paris which have the lowest average incomes per listing, at 7,016.02€ and 8,281.24€, respectively. Despite being major tourist destinations, the income per listing may be affected by either an oversupply of listings or lower nightly rates in competitive areas. Athens has a moderate income level, potentially indicating a balance between demand and supply. 
The city with the highest average nights per Listing is Venice at 117.68 which suggests consistent demand throughout the year,as a city with year round tourist interest. Athens, Lisbon, and Madrid also see relatively high booking averages, possibly due to their appeal as both tourist destinations and warmer climates, which can extend the travel season.
Copenhagen has the lowest average nights 30.59 despite high income per listing. This implies that listings in Copenhagen may experience high nightly rates, leading to substantial income with fewer bookings.
All in all from my analysis we can see that Athens does not present extreme values, neither very low nor very high, in any of the metrics that I examined which makes it particularly balanced. More specifically Athens shows a relatively high density of Airbnb listings compared to its population size. This high density its likely due to its popularity as a tourist destination.
For estimated nights and average income we can say that Athens has a high average for estimated nights per listing, meaning that properties are generally booked for longer durations than in some other cities. However, the average income per listing is lower compared to other cities, which may indicate fewer frequent bookings or economical problems. Nevertheless, this fact cannot be considered negative as it could reflect a more competative market, with more affordable accommodations, attracting a broader range of tourists.
So Athens is positioned as a competitive Airbnb market.

Athens has 381,932 total bookings and 1,145,796 total nights. This places Athens in a mid range position again, as we mentioned before, compared to other cities, indicating steady demand. Athens as a city with strong cultural and historical appeal, can easily benefit from promoting longer stays, something that will encourage travelers to immerse themselves in the local culture for extended periods.

```python
#data for licenses Visualisation
import numpy as np
dfr['license'] = dfr['license'].replace("NaN", np.nan) #Converting NaN to np.nan values
dfr['license_status'] = np.select([dfr['license'].notna() & (dfr['license'] != 0) & (dfr['license'] != 'Exempt'), #categories the values to license status
     dfr['license'].isna() | (dfr['license'] == 0), dfr['license'] == 'Exempt'],['Licensed', 'Unlicensed', 'Exempt'], default='Pending')
License = dfr.groupby(['City', 'license_status']).size().reset_index(name='Count')
total_listings_per_city = dfr.groupby('City').size().reset_index(name='Total_Listings')
License = License.merge(total_listings_per_city, on='City')
License['Proportion'] = License['Count'] / License['Total_Listings']
License
```

```python
#data for short term rentals Visualisation
rentals = dfr.groupby(['City', 'minimum_nights']).size().reset_index(name='Count')
total_listings_per_city = dfr.groupby('City').size().reset_index(name='Total_Listings')
rentals = rentals.merge(total_listings_per_city, on='City')
rentals['Proportion'] = rentals['Count'] / rentals['Total_Listings']
short_term_rentals = dfr[dfr['minimum_nights'] < 30]# Find rentals with nights below 30
long_term_rentals = dfr[(dfr['minimum_nights'] > 30) | (dfr['minimum_nights'] == 30)]# Find rentals with nights above 30
short_term_rentals = short_term_rentals.groupby('City').size().reset_index(name='Count_Below_30')# listings for each city with nights below 30
long_term_rentals = long_term_rentals.groupby('City').size().reset_index(name='Count_Above_30') #listings for each city with nights above 30
short_term_rentals = short_term_rentals.merge(long_term_rentals, on='City')
short_term_rentals = short_term_rentals.merge(total_listings_per_city, on='City')
short_term_rentals['Proportion_Below_30'] = short_term_rentals['Count_Below_30'] / short_term_rentals['Total_Listings']
short_term_rentals['Proportion_Above_30'] = long_term_rentals['Count_Above_30'] / short_term_rentals['Total_Listings']
rentals
```
